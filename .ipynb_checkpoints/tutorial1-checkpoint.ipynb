## Loading data into pandas

```python
import pandas as pd

#loading in csvs
df = pd.read_csv('pokemon_data.csv')
print(df.head(3))
```

```python
print(df.tail(3))
```

```python
#excel file
df_xlsx = pd.read_excel("pokemon_data.xlsx")
print(df_xlsx.head(3))
```

```python
#text file (separated by tabs)
df_txt = pd.read_csv("pokemon_data.txt", delimiter="\t")
print(df_txt)
```

## Reading Data in Pandas

```python
#reading headers 
print(df.columns)
```

```python
#Read each column
print(df['Name'])
```

```python
#multiple columns
print(df[['Name', 'Type 1', 'HP']])
```

```python
#specific row (integer location)
print(df.iloc[1:4])
```

```python
#specific location
#row,col
print(df.iloc[2, 1])
```

```python
#iterate through rows of a df
counter = 0
for index, row in df.iterrows():
    if counter < 5:
        print(index,row['Name'])
    counter += 1
```

```python
#finding data that isnt just based on index, conditionally show rows
df.loc[df['Type 1'] == 'Grass']
```

# Sorting/Describing Data

```python
#general information about the dataframe numerical columns
df.describe()
```

```python
#sort alphanumerically
df.sort_values('Name', ascending=False)
```

```python
#sort by two columns first one ascending and second one descending
#type one goes through A-Z and HP goes from high to low
df.sort_values(['Type 1', 'HP'], ascending=[True, False])
```

# Making changes to the data

```python
df.head(5)
```

```python
#adding new column
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
```

```python
#dropping a column
df = df.drop(columns=["Total"])
df.head(5)
```

```python
#add column in a new way
#axis=1 adding horizontally
#axis=0 adding vertically
df['Total'] = df.iloc[:,4:10].sum(axis=1)

df.head(5)
```

```python
#reordering columns
df_reordered = df[['Total', 'HP', 'Defense']]

df_reordered.head(5)
```

```python
#reordering columns elegantly
cols = list(df.columns.values)
df_reordered_2 = df[cols[0:4] + [cols[-1]] + cols[4:12]]

df_reordered_2.head(5)
```

# Saving Our Data (Exporting into desired format)

```python
#saves pandas indices too
df_reordered_2.to_csv("modified.csv")
#saves without indices
df_reordered_2.to_csv("modified_no_indicies.csv", index=False)
#save as excel file
df_reordered_2.to_excel('modified.xlsx', index=False)
#save tab separated file
df_reordered_2.to_csv('modified.txt', index=False, sep="\t")
```

# Filtering Data

```python
df = df_reordered_2
#equality operator on column
df.loc[df['Type 1'] == 'Grass']
```

```python
#more complex filtering conditions
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == "Poison")]
```
